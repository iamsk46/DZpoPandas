```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
data = pd.read_csv('kc_house_data.csv')
```

```python
data.info()
```

```python
def house_data(data):
    fig, axs = plt.subplots(1, 3, figsize=(15, 5));
    axs[0].hist(data['price'], bins=20, color='skyblue', edgecolor='black')
    axs[0].set_xlabel('Цена')
    axs[0].set_ylabel('Количество')
    axs[0].set_title('Распределение стоимости недвижимости')

    axs[1].hist(data['sqft_living'], bins=20, color='lightgreen', edgecolor='black')
    axs[1].set_xlabel('Жилая площадь')
    axs[1].set_ylabel('Количество')
    axs[1].set_title('Распределение жилой площади')

    axs[2].hist(data['yr_built'], bins=20, color='lightcoral', edgecolor='black')
    axs[2].set_xlabel('Год постройки')
    axs[2].set_ylabel('Количество')
    axs[2].set_title('Распределение года постройки')

    plt.tight_layout()
    plt.show()

house_data(data)
```

Z1.
Создал фунцкциюю для построения 3ех графиков. На первых 2ух графиках наблюдаю отрицательное скосное распределение. Основная масса значений распологается в первых ~25%. На третьем графке наблюдаю, что ,в основном, дома были построены во второй половине 20 Века. А в начале нового столетья случился "строительный бум", после чего значения стали приходить к средним показателям.

```python
def contrast(column):
    fig, axs = plt.subplots(3,2, figsize=(12, 8));
    for i, clm in enumerate(column):
        sns.countplot(x = data[clm], ax = axs[i][0])
        sns.boxplot(x = data[clm], y= data['price'], ax = axs[i][1])
    plt.setp(axs)
    plt.tight_layout()
    plt.show()

kol = ['floors', 'waterfront', 'condition']
contrast(kol)
```

Создал фунцкию для того, что бы сперва построить 3 графика с изучаемыми переменными(вид, этажность, состояние домов), затем эти переменные перенес на график "ящик с усами" и сравнил как они влияют на ценовой диапазон.
Из графиков вижу, что одноэтажные и двухэтажные дома встречаются чаще, чем остальные. Цена, в основном, повышается по мере увеличения количества этажей, но после 2,5 снижается.
Доля количества домов, имеющие выход к морю, очень мала. Поэтому цена в основном выше для домов, у которых выход к морю присутствует.
По мере увеличения уровня оценки состояния домов - цена также растет.
