```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
data = pd.read_csv('kc_house_data.csv')
```

```python
data.info()
```

```python
def house_data(data):
    fig, axs = plt.subplots(1, 3, figsize=(15, 5));
    axs[0].hist(data['price'], bins=20, color='skyblue', edgecolor='black')
    axs[0].set_xlabel('')
    axs[0].set_ylabel('')
    axs[0].set_title('  ')

    axs[1].hist(data['sqft_living'], bins=20, color='lightgreen', edgecolor='black')
    axs[1].set_xlabel(' ')
    axs[1].set_ylabel('')
    axs[1].set_title('  ')

    axs[2].hist(data['yr_built'], bins=20, color='lightcoral', edgecolor='black')
    axs[2].set_xlabel(' ')
    axs[2].set_ylabel('')
    axs[2].set_title('  ')

    plt.tight_layout()
    plt.show()

house_data(data)
```

```python
def contrast(column):
    fig, axs = plt.subplots(3,2, figsize=(12, 8));
    for i, clm in enumerate(column):
        sns.countplot(x = data[clm], ax = axs[i][0])
        sns.boxplot(x = data[clm], y= data['price'], ax = axs[i][1])
    plt.setp(axs)
    plt.tight_layout()
    plt.show()

kol = ['floors', 'waterfront', 'condition']
contrast(kol)
```
